```python
import pygame as pg
import random
```

# Basic use of pygame
In this section I draw some figures in a empty window

```python
pg.init()
size = (800, 500)

black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 0, 255) 
screen = pg.display.set_mode(size)
run = True
while run:
    
    screen.fill(white)
    pg.draw.line(screen, green, [0, 100], [200, 300], 5)
    pg.draw.rect(screen, black, (100, 100, 80, 80))
    pg.draw.circle(screen, black, (200, 200), 30)
    for i in range(100, 700, 100):
        pg.draw.rect(screen, black, (i, 230, 50, 50))
        
    pg.display.flip()
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
```

# Basic animation
In this section I create a basic animation of a rectangle

```python
pg.init()
size = (800, 500)
screen = pg.display.set_mode(size)
clock = pg.time.Clock()
run = True

black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 0, 255) 


cord_x = 400
cord_y = 200

speed_x = 3
speed_y = 3

while run:
    
    #Logic
    if (cord_x > 720 or cord_x < 0):
        speed_x *= -1
    if (cord_y > 420 or cord_y < 0):
        speed_y *= -1
        
    cord_x += speed_x
    cord_y += speed_y
    
    #Drawing
    screen.fill(white)
    pg.draw.rect(screen, red, (cord_x, cord_y, 80, 80))
    
    # Upload and clock
    pg.display.flip()
    clock.tick(60)
    
    # Exit game
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
```

# Animation part 2
In this section I create an animation with particles like snow

```python
pg.init()
size = (800, 500)
screen = pg.display.set_mode(size)
clock = pg.time.Clock()
run = True

black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 0, 255) 


coor_list = []
for i in range(60):
    x = random.randint(0, 800)
    y = random.randint(0, 500)
    coor_list.append([x, y])
    
while run:
    
    screen.fill(white)
    
    #Logic
    for coor in coor_list:
        
        pg.draw.circle(screen, red, coor, 2)
        coor[1] += 1
        if coor[1] >= 500:
            coor[1] = 0        
    
    #Drawing
    
    
    # Upload and clock
    pg.display.flip()
    clock.tick(60)
    
    # Exit game
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
```

# Mouse movement
In this section I create animation with mouse's movement

```python
pg.init()
size = (800, 500)
screen = pg.display.set_mode(size)
clock = pg.time.Clock()
run = True

black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 0, 255) 

# othe configuration
pg.mouse.set_visible(0)

while run:
    
    screen.fill(white)
    
    #Logic
    x, y = pg.mouse.get_pos()
    
    #Drawing
    pg.draw.rect(screen, red, (x, y, 100, 100))
    
    # Upload and clock
    pg.display.flip()
    clock.tick(60)
    
    # Exit game
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
```

# Keyboard movement
In this section I create a move with keyboard

```python
pg.init()
size = (800, 500)
screen = pg.display.set_mode(size)
clock = pg.time.Clock()
run = True

black = (0, 0, 0)
white = (255, 255, 255)
green = (0, 255, 0)
red = (255, 0, 0)
blue = (0, 0, 255) 

# othe configuration
x = 10
y = 10

x_sp = 0
y_sp = 0

while run:
    
    screen.fill(white)
    
    #Logic
    x += x_sp
    y += y_sp
    
    #Drawing
    pg.draw.rect(screen, red, (x, y, 100, 100))
    
    # Upload and clock
    pg.display.flip()
    clock.tick(60)
    
    # Exit game
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
        if event.type == pg.KEYDOWN:
            if event.key == pg.K_LEFT:
                x_sp = -3
            if event.key == pg.K_RIGHT:
                x_sp = 3
        
        if event.type == pg.KEYUP:
            if event.key == pg.K_LEFT:
                x_sp = 0
            if event.key == pg.K_RIGHT:
                x_sp = 0
        
        if event.type == pg.KEYDOWN:
            if event.key == pg.K_UP:
                y_sp = -3
            if event.key == pg.K_DOWN:
                y_sp = 3
        
        if event.type == pg.KEYUP:
            if event.key == pg.K_UP:
                y_sp = 0
            if event.key == pg.K_DOWN:
                y_sp = 0
```
