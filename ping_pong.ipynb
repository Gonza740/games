```python
import pygame as pg
```

```python
pg.init()
size = (800, 600)
screen = pg.display.set_mode(size)
clock = pg.time.Clock()
run = True

# othe configuration
black = (0, 0, 0)
white = (255, 255, 255)
player_width = 19
player_height = 90

#Player one move
x1_cor = 50
y1_cor = 255
y1_sp = 0

#Player two move
x2_cor = 700
y2_cor = 255
y2_sp = 0

#ball move
ball_x = 400
ball_y = 300
ball_sp_x = 3
ball_sp_y = 3

while run:
    
    screen.fill(black)
    
    #Drawing
    player_one = pg.draw.rect(screen, white, (x1_cor, y1_cor, player_width, player_height))
    player_two = pg.draw.rect(screen, white, (x2_cor, y2_cor, player_width, player_height))
    ball = pg.draw.circle(screen, white, (ball_x, ball_y), 10)
    
    #Logic
    # Playeres movement
    if (y1_cor > 0 or y1_sp > 0) and (y1_cor < 510 or y1_sp < 0):
        y1_cor += y1_sp
    if (y2_cor > 0 or y2_sp > 0) and (y2_cor < 510 or y2_sp < 0):
        y2_cor += y2_sp
    #Ball movement
    if ball.colliderect(player_one) or ball.colliderect(player_two):
        ball_sp_x *= -1
        
    ball_x += ball_sp_x
    ball_y += ball_sp_y
    
    if ball_y > 600 or ball_y < 0:
        ball_sp_y *= -1
    if ball_x > 800 or ball_x < 0:
        ball_x = 400
        ball_y = 300
        ball_sp_x *= -1
        ball_sp_y *= -1
    
    
    
    # Upload and clock
    pg.display.flip()
    clock.tick(60)
    
    # Exit game
    for event in pg.event.get():
        if event.type == pg.QUIT:
            pg.quit()
            pg.display.quit()
            run = False
            break
        if event.type == pg.KEYDOWN:
            if  event.key == pg.K_w:
                y1_sp = -3
            if event.key == pg.K_s:
                y1_sp = 3
            if event.key == pg.K_UP:
                y2_sp = -3
            if event.key == pg.K_DOWN:
                y2_sp = 3
        
        if event.type == pg.KEYUP:
            if event.key == pg.K_w:
                y1_sp = 0
            if event.key == pg.K_s:
                y1_sp = 0
            if event.key == pg.K_UP:
                y2_sp = 0
            if event.key == pg.K_DOWN:
                y2_sp = 0
```
